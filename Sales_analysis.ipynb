```python
# Data Visualisation
```

```python
monthly_revenue=df.groupby("Month")["Revenue"].sum()
monthly_revenue
```

```python
plt.plot(monthly_revenue.index,monthly_revenue.values,color='red',marker="o",label="Monthly sales")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Month vs Revenue")
plt.legend()
plt.grid(True,alpha=0.5)
plt.ticklabel_format(style="plain", axis="y")
plt.savefig("month_vs_revenue.png")
plt.show()
```

```python
product_revenue=df.groupby("Product")["Revenue"].sum()
product_revenue
```

```python
plt.bar(product_revenue.index,product_revenue.values,color="orange")
plt.xlabel("Products")
plt.ylabel("Revenue")
plt.title("Product revenue")
plt.ticklabel_format(style="plain", axis="y")
plt.grid(axis="y",alpha=0.5)
plt.savefig("product_vs_revenue.png")
plt.show()
```

```python
revenue_share=df.groupby("Product")["Revenue"].sum()
revenue_share
```

```python
explode=[0.1,0.1,0]
plt.pie(revenue_share.values,labels=revenue_share.index,colors=["Red","Lime","skyblue"],autopct="%1.1f%%",startangle=60,
    shadow=True,
    explode=explode)
plt.title("Revenue share per product")
plt.savefig("revenue_share_pie.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Revenue"],bins=6,edgecolor="black",color="skyblue")

plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.ticklabel_format(style="plain", axis="x")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.savefig("revenue_distribution_histogram.png")
plt.show()
```

```python
unitssold=df.groupby("Units_Sold")["Revenue"].sum()
plt.scatter(unitssold.index,unitssold.values)
plt.xlabel("Units Sold")
plt.ylabel("Revenue")
plt.title("Units sold vs Revenue")
plt.grid(True,color="gray")
plt.ticklabel_format(style="plain",axis="y")
plt.savefig("units_sold_vs_revenue_scatter.png")
plt.show()
```

```python
# ===== Sales Analysis Report =====

# Total Revenue:
print("Total Revenue:",df["Revenue"].sum())

# Total Units Sold:
print("Total Units Sold:",df["Units_Sold"].sum())

# Highest & Lowest Revenue Product:
rev_data=df.groupby("Product")["Revenue"].sum()
print("Product with Highest Revenue:",rev_data.idxmax())
print("Product with Lowest Revenue:",rev_data.idxmin())

# Highest & Lowest Revenue Month:
rev_month=df.groupby("Month")["Revenue"].sum()
print("Highest Revenue Month:",rev_month.idxmax())
print("Lowest Revenue Month:",rev_month.idxmin())

# Average Revenue:
print("Average Revenue:",df["Revenue"].mean())

# Highest Selling Product:
highest_selling=df.groupby("Product")["Units_Sold"].sum()
print("Highest Selling Product:",highest_selling.idxmax())

# Total Electronics Revenue:
print("Total Electronics Revenue:",df[df["Category"]=="Electronics"]["Revenue"].sum())

# Total Furniture Revenue:
print("Total Furniture Revenue:",df[df["Category"]=="Furniture"]["Revenue"].sum())
```
